```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time 
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
driver.get("https://admin-dev.omadar.finance/farms/")
time.sleep(5)
```

```python
driver.find_element_by_xpath("//div[@class='ant-row ant-row-end']//button[@class='ant-btn ant-btn-primary']").click()
time.sleep(1)
```

```python
# Mastercheft Input
driver.find_element_by_id('manageFarms_masterchefAddress').send_keys(Keys.COMMAND + 'a')
time.sleep(1)
driver.find_element_by_id('manageFarms_masterchefAddress').send_keys(Keys.DELETE)
time.sleep(1)
driver.find_element_by_id("manageFarms_masterchefAddress").send_keys("0x671e56C68047029F236f342b18632425C75885a3")
```

```python
# Protocol Input
driver.find_element_by_id('manageFarms_title').send_keys(Keys.COMMAND + 'a')
time.sleep(1)
driver.find_element_by_id('manageFarms_title').send_keys(Keys.DELETE)
time.sleep(1)
driver.find_element_by_id("manageFarms_title").send_keys("Honeyfarm")
time.sleep(1)
```

```python
# Pool ID Input
driver.find_element_by_id("manageFarms_poolId").send_keys("47")
time.sleep(1)
```

```python
# LP Input
driver.find_element_by_id("manageFarms_lpAddress").send_keys("0xdDdc245c62a0AA875893Ab53b52455aCdCd7f526")
time.sleep(1)
```

```python
# Chain Select
driver.find_element_by_id("manageFarms_chainId").click()
driver.find_element_by_xpath("//div[@class='ant-select-item-option-content' and contains(text(),'BSC')]").click()
time.sleep(1)
```

```python
# Dropdown Select
def drpSelect(reward, count, tag, tagList):
    driver.find_element_by_id(tag).click()
    time.sleep(1)
    for i in range(count):
        if driver.find_elements_by_xpath("//div[@id='%s']//div[@aria-label='%s']" % (tagList, reward)):
            driver.find_element_by_id(tag).send_keys(Keys.DOWN, Keys.ENTER)
            time.sleep(1)
            break
        else: 
            driver.find_element_by_id(tag).send_keys(Keys.DOWN)
            time.sleep(0.1)

drpSelect('MOON', 278, 'manageFarms_rewardToken', 'manageFarms_rewardToken_list')
drpSelect('SING', 278, 'manageFarms_ref1', 'manageFarms_ref1_list')
drpSelect('BUSD', 278, 'manageFarms_ref2', 'manageFarms_ref2_list')
```
