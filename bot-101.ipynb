```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
import time 
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
driver.get("https://admin-dev.omadar.finance/farms/")
time.sleep(5)
```

```python
driver.find_element_by_xpath("//div[@class='ant-row ant-row-end']//button[@class='ant-btn ant-btn-primary']").click()
time.sleep(1)
```

```python
# Mastercheft Input
driver.find_element_by_id('manageFarms_masterchefAddress').send_keys(Keys.COMMAND + 'a')
time.sleep(1)
driver.find_element_by_id('manageFarms_masterchefAddress').send_keys(Keys.DELETE)
time.sleep(1)
driver.find_element_by_id("manageFarms_masterchefAddress").send_keys("0x0eD7e52944161450477ee417DE9Cd3a859b14fD0")
```

```python
# Protocol Input
driver.find_element_by_id('manageFarms_title').send_keys(Keys.COMMAND + 'a')
time.sleep(1)
driver.find_element_by_id('manageFarms_title').send_keys(Keys.DELETE)
time.sleep(1)
driver.find_element_by_id("manageFarms_title").send_keys("PancakeSwap")
time.sleep(1)
```

```python
# Pool ID Input
driver.find_element_by_id("manageFarms_poolId").send_keys("251")
time.sleep(1)
```

```python
# LP Input
driver.find_element_by_id("manageFarms_lpAddress").send_keys("0x0eD7e52944161450477ee417DE9Cd3a859b14fD0")
time.sleep(1)
```

```python
# Chain Select
driver.find_element_by_id("manageFarms_chainId").click()
driver.find_element_by_xpath("//div[@class='ant-select-item-option-content' and contains(text(),'BSC')]").click()
time.sleep(1)
```
